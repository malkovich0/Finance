```python
import pandas as pd
import os
import re
import numpy as np
```

```python
data_path = './Raw/fssdata/'
files_list = os.listdir(data_path)
```

```python
file = data_path+'005930.xlsx'
```

```python
value_dict = {'총자산':'ifrs-full_Assets', '현금':'ifrs-full_CashAndCashEquivalents',
              '부채':'ifrs-full_Liabilities','지배자산':'ifrs-full_EquityAttributableToOwnersOfParent',
             '매출액':'ifrs-full_Revenue', '매출원가':'ifrs-full_CostOfSales', 
              '매출총이익':'ifrs-full_GrossProfit', '판관비':'dart_TotalSellingGeneralAdministrativeExpenses',
              '영업이익':'dart_OperatingIncomeLoss','계속영업이익':'ifrs-full_ProfitLossFromContinuingOperations',
              '당기순이익':'ifrs-full_ProfitLoss', '지배순이익':'ifrs-full_ProfitLossAttributableToOwnersOfParent',
             '영업활동현금흐름':'ifrs-full_CashFlowsFromUsedInOperatingActivities'}
```

```python
value_sheet_dict = {'Data_bs': ['총자산','현금','부채','지배자산'],
                    'Data_is':['매출액','매출원가','매출총이익','판관비','영업이익',
                               '계속영업이익','당기순이익','지배순이익'],
                    'Data_cis':[],
                    'Data_cf':['영업활동현금흐름']}
```

```python
sheet_list = ['Data_bs','Data_is','Data_cis','Data_cf']
```

```python
sheet_name = sheet_list[0]
excel_temp = pd.read_excel(file, sheet_name=sheet_name,index_col=0)

columns = ['stock_code','period','report_name']+list(value_dict.keys())
df_temp_total = pd.DataFrame(columns=columns)

columns_new = []
p = re.compile('[0-9]+')
for i in excel_temp.columns:
    m = p.match(i)
    if not m == None:
        columns_new.append(m.group()) #group() : 매치된 문자열 반환
df_temp_total['period']=columns_new

for value_name in value_sheet_dict[sheet_name]:
    df_temp = excel_temp[excel_temp[excel_temp.columns[0]] == value_dict[value_name]]
    df_temp = df_temp[columns_new].T
    df_temp_list = df_temp.iloc[:,0].to_list()
    df_temp_total[value_name]=df_temp_list
    
df_temp_total['stock_code']='005930'
```

```python
columns_new_start = [col.split('-')[0][:6] for col in columns_new]
columns_new_end = [col.split('-')[1][:6] for col in columns_new]
```

```python
columns_new_end[0]
```

```python
col_start = columns_new_start[0]
col_end = columns_new_end[0]
```

```python
def find_index(data, target):
    res = []
    lis = data
    while True:
        try:
            res.append(lis.index(target) + (res[-1]+1 if len(res)!=0 else 0))
            lis = data[res[-1]+1:]
        except:
            break     
    return res
```

```python
col_start[:4]+'01'
```

```python
add_list = []
add_list1 = []
del_list = []

for i in range(len(columns_new_start)):
# range(2):

    col_start = columns_new_start[i]
    col_end = columns_new_end[i]

    if col_start[4:] == '01':
        add_list.append(np.NaN)
    else:
        try:
            end_target_index = find_index(columns_new_end, str(int(col_start)-1)) #이번 기간의 전 기간에서 끝나는 index
            start_target_index = find_index(columns_new_start, col_start[:4]+'01') #해당 연도의 1월부터 시작하는 index
            intersection = list(set(end_target_index) & set(start_target_index)) #교집합을 더하면 1월부터 col_end까지 계산가능
            # intersection이 없으면 error를 더하게 수정
            add_list.append(intersection)
#             print(intersection)
        except:
            add_list.append('Error')
            
        # 전해의 전체 값 중 필요 기간을 제외하고 빼기

    if col_end[4:] == '12':
        add_list1.append(np.NaN)
        del_list.append(np.NaN)
    else:
        try:
#             print(str(int(col_end[:4])-1)+'12')
            end_target_index_add = find_index (columns_new_end, str(int(col_end[:4])-1)+'12' )
            end_target_index_del = find_index (columns_new_end, str(int(col_end[:4])-1)+col_end[4:] )
            start_target_index = find_index(columns_new_start, str(int(col_end[:4])-1)+'01' )
            intersection_add = list(set(end_target_index_add) & set(start_target_index))
            intersection_del = list(set(end_target_index_del) & set(start_target_index))
            add_list1.append(intersection_add)
            del_list.append(intersection_del)
        except:
            add_list1.append('Error')
            del_list.append('Error')
```

```python
print(len(add_list), len(add_list1), len(del_list))
```
