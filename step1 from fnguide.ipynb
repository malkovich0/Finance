```python
def corp_data (stock_code):
    columns = ['stock_code','period','총자산','현금','부채','지배자산',
           '매출액','매출원가','매출총이익','판관비','영업이익','당기순이익','지배순이익','영업활동현금흐름']
    df_temp = pd.DataFrame(columns=columns)
    url_fin = 'http://comp.fnguide.com/SVO2/ASP/SVD_Finance.asp?pGB=1&gicode=A{}&cID=&MenuYn=Y&ReportGB=D&NewMenuID=103&stkGb=701'.format(stock_code)
    tables = pd.read_html(url_fin)

    cis = tables[0]
    cis = cis.set_index(keys=cis.columns[0])
    bs = tables[2]
    bs = bs.set_index(keys=bs.columns[0])
    cf = tables[4]
    cf = cf.set_index(keys=cf.columns[0])

    cols = bs.columns[0:3]
    cols_modi = [col[0:4]+col[-2:] for col in cols]

    df_temp['총자산'] = bs.loc['자산',cols]
    df_temp['부채'] = bs.loc['부채',cols]
    df_temp['지배자산'] = bs.loc['비지배주주지분',cols]
    df_temp['지배자산'].fillna(0,inplace=True)
    df_temp['지배자산'] = bs.loc['자본',cols] - df_temp['지배자산']

    if '매출액' in cis.index:
        df_temp['매출액'] = cis.loc['매출액',cols]
#     if '순이자손익' in cis.index:
#         df_temp['매출액'] = cis.loc['순이자손익',cols]
    if '매출원가' in cis.index:
        df_temp['매출원가'] = cis.loc['매출원가',cols]
    if '매출총이익' in cis.index:
        df_temp['매출총이익'] = cis.loc['매출총이익',cols]
    # df_temp['판관비'] = cis.loc['',cols]
    df_temp['영업이익'] = cis.loc['영업이익',cols]
    df_temp['당기순이익'] = cis.loc['당기순이익',cols]
    df_temp['지배순이익'] = cis.loc['비지배주주순이익',cols]
    df_temp['지배순이익'].fillna(0,inplace=True)
    df_temp['지배순이익'] = cis.loc['당기순이익',cols] - df_temp['지배순이익']

    df_temp['영업활동현금흐름'] = cf.loc['영업활동으로인한현금흐름',cols]

    df_temp['period'] = cols_modi
    df_temp['stock_code'] = stock_code
    return df_temp
```
