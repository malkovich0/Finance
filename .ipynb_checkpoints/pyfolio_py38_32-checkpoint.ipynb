```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Product/Momentum_Cap_5partial_11m_est_w_1m_gap_3m_selec_1m_rebal.csv',index_col=0)
df.cumprod()
```

```python
def date_to_date (date):
    date = date.split('-')
    date_new = datetime(int(date[0]),int(date[1]),int(date[2]))
    return date_new
```

```python
df.index = [date_to_date(date) for date in df.index ]
```
